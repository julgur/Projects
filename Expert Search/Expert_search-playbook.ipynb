## Appendix
here there are some experiments on data to become sure the functions work correctly. Due to the differences of the types of fields, some missings may be filled with random stuff instead of None. Like string 'NaN', empty lists, and so on. Here we count how many measings are not got by ifs in the main function

```python
counter = 0
for art in articles:
    cites = art['keywords']
    if type(cites)==str and cites!='NaN':
        print(cites)
        counter +=1
print(counter, len(articles), counter/len(articles))  
```

```python
counter = 0
for art in articles:
    cites = art['workLinks']
    if cites==[] or cites == None:
        counter +=1
print(counter, len(articles), counter/len(articles))        
```

```python
counter = 0
for art in articles:
    cites = art['authorID']
    if cites==[] or cites == None or type(cites)!=list:
        counter +=1
print(counter, len(articles), counter/len(articles))        
```
