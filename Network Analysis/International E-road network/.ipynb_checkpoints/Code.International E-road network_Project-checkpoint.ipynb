 ## Degree distribution. 
Now we study degree distribution. 

```python
deg = dict(nx.degree(G))
sorted_nodes = [(k,v) for k, v in deg.items()]
sorted_nodes=sorted(sorted_nodes, key= lambda x: x[1], reverse = True)

for k,f in sorted_nodes[:20]:
    print(k, nx.get_node_attributes(G,'name')[k], f)
```

```python
for j in G.neighbors(284):
  print(j, nx.get_node_attributes(G,'name')[j])
```

```python
degree_sequence = sorted(dict(G.degree()).values(), reverse=True)
degreeCount = collections.Counter(degree_sequence)
k=sorted(degreeCount.keys())
s=sum(degreeCount.values())
pdf=[degreeCount[i]/s for i in k]


fig = plt.figure(figsize=(10,6))
plt.plot(k, pdf,'o')
plt.title('PDF for degree distribution')
plt.xlabel('k')
plt.ylabel('prob of having degree k')
```

```python
components = list(nx.connected_component_subgraphs(G))
print('number of connected components', len(components))
max_comp= max(nx.connected_component_subgraphs(G), key=len)
print('size of largest connected component', len(max_comp))
```

```python

print('number of nodes:', nx.number_of_nodes(G))
print('average degree:', np.mean(degree_sequence))
print('diameter:', nx.diameter(max_comp))
print('average shortest path length:', nx.average_shortest_path_length(max_comp))
print('clustering coefficient (transitivity):',nx.transitivity(G))
```

```python
degree = nx.algorithms.centrality.degree_centrality(G)
eigenvector =  nx.eigenvector_centrality_numpy(G)
closeness = nx.closeness_centrality(max_comp)
btw = nx.betweenness_centrality(G)
```

```python
def print_highest(degree):
    highest_degree = [(k,v) for k, v in dict(degree).items()]
    highest_degree=sorted(highest_degree, key= lambda x: x[1], reverse = True)
    for k,a in highest_degree[:20]:
        print(k, nx.get_node_attributes(G,'name')[k], "{0:.4f}".format(a))
        

def high_graph(deg, t, graph, save_name, nodesize=1000):
    degree = dict(deg)
    filtered_nodes = [k for k, v in degree.items() if v > t]
    subgraph = nx.subgraph(graph, filtered_nodes)

    
    rang = [degree[i] for i in subgraph.nodes()]

#    degree = nx.degree(subgraph)
    labels = nx.get_node_attributes(subgraph, 'name')

    plt.figure(1, figsize=(15,7))
    plt.axis('off')
    pos = nx.spring_layout(subgraph, k=5/np.sqrt(subgraph.order()))
    nx.draw_networkx(subgraph,
                     pos=pos,
                     node_size = nodesize,
                     node_color=rang,
                     font_size=10,
                     font_color='red',
                     cmap=plt.cm.Blues,
                     with_labels=True,
                     labels=labels,
                     edge_color='lightgrey',
                     alpha=1)
    plt.savefig(save_name)
```

```python
print('The nodes with highest degree centrality')
print_highest(degree)
```

```python
high_graph(degree, 0.005,G,'degree.png',nodesize = 2000)
```

```python
print('The nodes with highest closeness centrality')
print_highest(closeness)
```

```python
high_graph(closeness, 0.0812,G,'close.png',nodesize=2000)
```

```python
print('The nodes with highest Betweenness centrality')
print_highest(btw)
```

```python
high_graph(btw, 0.1,G,'btw.png', nodesize=1000)
```

```python
print('The nodes with highest eigenvector centrality')
print_highest(eigenvector)
```

```python
high_graph(eigenvector, 0.1,G,'eig.png', nodesize=1000)
```

```python
pr = nx.pagerank(G, alpha=0.8)
print('The nodes with highest Page-Rank centrality')
print_highest(pr)
```

```python
high_graph(pr, 0.0015,G,'pr.png', nodesize=1000)
```

```python
communities_generator = community.k_clique_communities(G,2)
clique_communities=list(communities_generator)
clique_communities
```

```python
clique_color={}
i=2
for c in clique_communities:
    for node in c:
        clique_color[node] = i
    i+=1
for n in G.nodes():
    clique_color[n]=clique_color.get(n,1)*3
len(clique_color) == G.number_of_nodes()
```

```python
plt.figure(1, figsize=(10,7))
plt.axis('off')
pos = nx.spring_layout(G)
labels = nx.get_node_attributes(G, 'name')
nx.draw_networkx(G,
                     pos=pos,
                     node_size = 500,
                     node_color=[clique_color[n] for n in G.nodes()],
                     font_size=9,
                     font_color='red',
                  #   cmap=plt.cm.Blues,
                     with_labels=False,
                    labels=labels,
                    edge_color='lightgrey',
                     alpha=0.8
                )
plt.savefig('cliques.png')
```
